```python
# plot problem 12
import matplotlib.pyplot as plt
import numpy as np

data = np.genfromtxt('result12.csv', delimiter=',')

group = ["0.01", "0.1", "1", "10", "100"]
gamma_count = [np.sum(data[:,1] == float(g)) for g in group]

plt.bar(group, gamma_count, width=0.5, color='#5448C8')
plt.xlabel('gamma')
plt.ylabel('count')
plt.title('Problem 12 best gamma')
plt.savefig('p12.png')
```

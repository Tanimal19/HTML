```python
# plot problem 10
import matplotlib.pyplot as plt
import numpy as np

data = np.genfromtxt('result10.csv', delimiter=',')
eouts = data[:,1]
non_zeros = data[:,2]

# histogram of eouts
plt.figure()
plt.hist(eouts, color='#5448C8')
plt.xlabel('Eout')
plt.ylabel('Frequency')
plt.title('Problem 10 Eout Histogram')
plt.savefig('problem10_eout_hist.png')

# histogram of non-zeros
plt.figure()
plt.hist(non_zeros, color='#5448C8')
plt.xlabel('Non-zero components')
plt.ylabel('Frequency')
plt.title('Problem 10 Non-zero Components Histogram')
plt.savefig('problem10_non_zeros_hist.png')
```
